##Parte 1: Trabajo con matrices usando SymPy

1. Primero importamos la librería.

```python
import sympy as sp
```

2. En SymPy podemos crear matrices usando Matrix.


```python
A = sp.Matrix([[4, 7], [-5, 0]])
S = sp.Matrix([[8, 1], [3, 6]])

print(A)
print() #Espacio para que no se vean demasiado juntas las matrices
print(S)
```

<div style="text-align: justify">Para mejorar la presentación visual, podemos importar la función pprint de la librería SymPy, la cual permite mostrar las matrices en un formato más claro y similar a la notación matemática convencional.</div>


```python
from sympy import pprint  #Así es como se importa
M = sp.Matrix([[4, 7], [-5, 0]])
N = sp.Matrix([[8, 1], [3, 6]])

pprint(M)  #Esta es la que hace que se "impriman de mejor manera", visualmente hablando.
print()
pprint(N)  #Esta es la que hace que se "impriman de mejor manera", visualmente hablando.
```

3. Obtener una entrada específica de la matriz


<div style="text-align: justify">Para acceder a una entrada específica usamos índices, recordemos que en Python los índices comienzan en 0. La sintaxis es:</div> $$\text{Matriz}[\text{fila},\text{columna}]$$

```python

A[0,1] #Como se inicia en 0, esto dice la primera fila y la segunda columna
```

Esto devuelve la entrada en la fila 1, columna 2 de mi matriz $A$

4. Obtener un renglón

Para obtener un renglón usamos el método row().

```python
A.row(0)
```

Esto devuelve el primer renglón de mi matriz $A$.

5. Obtener una columna

Para obtener una columna usamos el método col().


```python
A.col(1)
```

Esto devuelve la segunda columna de mi matriz $S$.

6. Multiplicación de matrices

La multiplicación de matrices se realiza con el operador *.

Recordemos que para multiplicar matrices:

$$A_{mxn} \cdot S_{nxp}$$
	​


las dimensiones internas deben coincidir.

```python
A*S
```

7. Elevar una matriz a una potencia

<div style="text-align: justify">Podemos elevar una matriz a una potencia usando el operador **. Esto equivale a multiplicar la matriz por sí misma varias veces.</div>

```python
A**2
```

```python
S**3
```

8. Multiplicar un vector por una matriz

Un vector puede escribirse como una lista de números. Por ejemplo, el vector
$$v = (1,2)$$

En SymPy, los vectores se representan como matrices columna.

```python
# Definimos el vector como matriz columna
v = sp.Matrix([1, 2])

# Mostramos
pprint(v)
```

Ahora podemos multiplicar la matriz $A$ por el vector $v$:

```python
A*v
```
