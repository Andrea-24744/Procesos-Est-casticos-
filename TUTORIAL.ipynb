##Parte 1: Trabajo con matrices usando SymPy

**1. Primero importamos la librería.**

```python
import sympy as sp
```

**2. En SymPy podemos crear matrices usando Matrix.**


```python
A = sp.Matrix([[4, 7], [-5, 0]])
S = sp.Matrix([[8, 1], [3, 6]])

print(A)
print() #Espacio para que no se vean demasiado juntas las matrices
print(S)
```

<div style="text-align: justify">Para mejorar la presentación visual, podemos importar la función pprint de la librería SymPy, la cual permite mostrar las matrices en un formato más claro y similar a la notación matemática convencional.</div>


```python
from sympy import pprint  #Así es como se importa
M = sp.Matrix([[4, 7], [-5, 0]])
N = sp.Matrix([[8, 1], [3, 6]])

pprint(M)  #Esta es la que hace que se "impriman de mejor manera", visualmente hablando.
print()
pprint(N)  #Esta es la que hace que se "impriman de mejor manera", visualmente hablando.
```

**9. Valores propios de una matriz**

Los valores propios (eigenvalues) de una matriz son números que satisfacen:
$$Sv=λv$$

donde:
* $S$ es la matriz
* $v$ es un vector propio
* $λ$ es un valor propio

En SymPy se obtienen con eigenvals().

```python
S.eigenvals()
```

se interpreta de la siguiente manera:

* Las claves del diccionario (los números antes de los dos puntos) son los valores propios (eigenvalues) de la matriz.
* Los valores del diccionario (los números después de los dos puntos) indican la multiplicidad algebraica de cada valor propio, es decir, cuántas veces aparece ese valor propio.

significa que:

* 9 es un valor propio de la matriz y aparece 1 vez.
* 5 es otro valor propio de la matriz y aparece 1 vez.

Por lo tanto, la matriz S tiene dos valores propios diferentes: 9 y 5, y cada uno tiene multiplicidad 1.

**10. Vectores propios**

Para obtener valores propios y vectores propios usamos eigenvects().



```python
S.eigenvects()
```

La salida incluye:
$$(\text{valor_propio}, \text{multiplicidad}, \text{[vectores_propios]})$$

Es decir:

* Primer elemento: el valor propio **$(λ)$**
* Segundo elemento: la **multiplicidad** algebraica
* Tercer elemento: una lista con los vectores propios asociados
**texto en negrita**

Se puede imprimir mejor para que sea más entendible:

```python
for val, mult, vects in S.eigenvects():
    print("Valor propio:", val)
    print("Multiplicidad:", mult)
    print("Vector(es) propio(s):")
    for v in vects:
        print(v)
    print()
```

## Parte 2: Diagramas de transición con NetworkX

> *¿Qué es un diagrama de transición?*

Un diagrama de transición representa un sistema compuesto por estados y transiciones entre estados. Se modela mediante un grafo dirigido, donde:

* Los nodos representan estados del sistema.
* Las aristas dirigidas representan transiciones entre estados.

Para construir y visualizar este tipo de diagramas en Python utilizaremos la librería NetworkX, junto con Matplotlib para la visualización.

1. Importar librerías

```python
import networkx as nx
import matplotlib.pyplot as plt   # Importamos las librerias necesarias
```

2. Crear un grafo dirigido

Un diagrama de transición necesita dirección en las transiciones, por lo que usamos DiGraph.

```python
G = nx.DiGraph()
```

Aquí estamos creando un grafo dirigido vacío, al cual posteriormente le agregaremos estados y transiciones.

3. Agregar estados(nodos)

Los estados sistema se representan como nodos en el grafo.

```python
G.add_nodes_from(["A","B","C"])
```

Esto crea tres estados:

* A
* B
* C

Cada uno representa una posible configuración del sistema.

4. Agregar transiciones

Las transiciones entre estados se representan mediante aristas dirigidas.

```python
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","A")
G.add_edge("A","C")
```

Esto significa:

* El sistema puede pasar de $A → B$
* De $B → C$
* De $C → A$
* También directamente de $A → C$

Cada arista indica un posible cambio de estado.

5. Dibujar el diagrama

Para visualizar el grafo usamos networkx.draw().

```python
pos = nx.spring_layout(G)

nx.draw(G,pos, with_labels=True,
        node_color="lightblue",
        node_size=2000,
        font_size=12,
        font_weight="bold",
        arrows=True)

plt.show()
```

Explicación de los parámetros

* *spring_layout(G):* Calcula automáticamente una posición equilibrada de los nodos para que el grafo sea fácil de leer.
* *with_labels=True*: Muestra el nombre de cada estado en el nodo.
* *node_color*: Define el color de los estados.
* *node_size*: Controla el tamaño de los nodos.
* *font_size*: Tamaño del texto de los estados.
* *arrows=True*: Muestra flechas, indicando la dirección de las transiciones.

Interpretación del diagrama

El diagrama generado muestra:

* Estados representados por círculos.
* Flechas que indican las posibles transiciones.
* La dirección de cada flecha muestra hacia qué estado se mueve el sistema.

Por ejemplo:

* Si el sistema está en A, puede pasar a B o C.
* Si está en B, puede pasar a C.
* Si está en C, puede regresar a A.
