# **Simulamos la funciòn volado**
$$  f(x) = \begin{cases}
    \text{Sol} & \text{si } x < 0.5 \\
    \text{Àguila} & \text{si } x \geq 0.5
\end{cases} $$

```python
def volado():
  x = r.random()
  if x >= 0  and x < 0.5:
    return "Sol"
  else:
    return "Àguila"
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np
def voladoAgui(n):
    aguila = sum(1 for _ in range(n) if random.random() >= 0.5)  # 50% de águila
    return aguila / n  # Probabilidad de obtener águila

def experimento(m):
    """Realiza m experimentos y almacena la probabilidad obtenida en cada uno."""
    p = []
    for j in range(1, m + 1):  # Empezar desde 1 hasta m
        p.append(voladoAgui(j))  # Llamar correctamente a volado y guardar el resultado
    return p

# Parámetro de número de experimentos
m = 9000  # Cambia esto si quieres más experimentos

# Ejecutar el experimento
p = experimento(m)

# Gráfico
plt.figure(figsize=(12, 4))
plt.plot(p,color ="#9BFAB0", linestyle='-')
plt.axhline(y=0.5, color="#3EB489", linestyle='--', label="Esperado (0.5)")
plt.title("Probabilidad generada de obtener águila")
plt.xlabel("Número de experimento")
plt.ylabel("P(Águila)")
plt.legend()
plt.grid(True)
plt.show()
print("La media es: ", np.mean(p))
print("La varianza es: ",np.var(p))
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np
def voladoSol(n):
    Sol = sum(1 for _ in range(n) if random.random() < 0.5)  # 50% de Sol
    return Sol / n  # Probabilidad de obtener Sol

def experimento(m):
    """Realiza m experimentos y almacena la probabilidad obtenida en cada uno."""
    z = []
    for j in range(1, m + 1):  # Empezar desde 1 hasta m
        z.append(voladoSol(j))  # Llamar correctamente a volado y guardar el resultado
    return z

# Parámetro de número de experimentos
m = 1000

# Ejecutar el experimento
z = experimento(m)

# Gráfico
plt.figure(figsize=(12, 5))
plt.plot(z,color ="skyblue", linestyle='-')
plt.axhline(y=0.5, color='b', linestyle='--', label="Esperado (0.5)")
plt.title("Probabilidad generada de obtener Sol")
plt.xlabel("Número de experimento")
plt.ylabel("Z(Sol)")
plt.legend()
plt.grid(True)
plt.show()
np.mean(z)
```

```python
import random as r
import numpy as np
import matplotlib.pyplot as plt

def moneda(n):
  if n==0:
    return 0.0
  sum = 0
  for _ in range(n):
     x = r.random()
     if x < 0.5:
      sum += 1
  return sum/n
```

```python
def repeticion(m):
  p = []
  for i in range(m):
    p.append(moneda(i))
  return p
```

```python
d = repeticion(1000)
print("La media es: ", np.mean(d))
print("La varianza es: ",np.var(d))
plt.plot(d)
plt.plot(d,color ="skyblue", linestyle='-')
plt.axhline(y=0.5, color='b', linestyle='--', label="Esperado (0.5)")
plt.legend()
#plt.grid(True)
plt.show()
```
